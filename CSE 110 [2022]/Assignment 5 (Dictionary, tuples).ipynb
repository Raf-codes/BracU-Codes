```python
user_input1 = input("Enter a dictionary without brackets'': ").split(",")
dictionary1 = {}

for item in user_input1:
  dictionary1.update({item.split(":")[0].strip() : item.split(":")[-1].strip()})

count = 0
sum = 0

for value in dictionary1.values():
  count += 1
  sum += int(value)

print("Average is " + str(int(sum/count)) + ".")
```
