```python
#dict2
primaryDict =dict()
while True:
    in_num = input()
    if in_num.upper() == "STOP":
        break
    else:
        num = int(in_num)
        if num in primaryDict.keys():
            primaryDict[num] += 1
        else:
            primaryDict[num] = 1

for key,value in primaryDict.items():
    print(key,"-", value, " times")
```

```python
#fun5
def Palindrome(s):
    a = ""
    for i in s:
        if i != " ":
            a += i


    b = ""
    for i in range(len(s)-1,-1,-1):
        if s[i] != " ":
            b += s[i]
    if a == b:
        print("Palindrome")
    else:
        print("Not a Palindrome")


Palindrome("nurses run")
Palindrome("hello")
```

```python
# function 5
def palindrome(s):
    a = ''
    for i in s:
        if i != ' ':
            a += i
    b=len(s)
    if a == (a[b::-1]):
        print ("Palindrome")
    else:
        print ("Not a palindrome")


palindrome ("nurses run")
palindrome("hello")
```

```python
#function6
def capitalization(s):
    #print(s)
    l = len(s)
    out = ""

    for i in range(l):
        if i == 0:

            if ord("a") <= ord(s[i]) <= ord("z"):
                new = ord(s[i]) - 32
                out += chr(new)

            else:
                out += s[i]

        elif i > 2 and (s[i - 2] == "?" or s[i - 2] == "." or s[i - 2] == "!"):
            if ord("a") <= ord(s[i]) <= ord("z"):
                new = ord(s[i]) - 32
                out += chr(new)
            else:
                out += s[i]

        elif s[i] == "i":
            if s[i - 1] == " " and (
                    s[i + 1] == " " or s[i + 1] == "." or s[i + 1] == "?" or s[i + 1] == "!"):
                new = ord(s[i]) - 32
                out += chr(new)

            else:
                out += s[i]

        else:
            if ord("A") <= ord(s[i]) <= ord("Z"):
                new = ord(s[i]) + 32
                out += chr(new)
            else:
                out += s[i]
    print(out)


capitalization('my favourite animal is a dog. a dog has sharp teeth so that it can eat flesh very easily. do you know my pet dog’s name? i love my pet very much.')
```
