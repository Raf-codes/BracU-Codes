```python
#25


class Vaccine:
    
    def __init__(self, name, country, days):
        self.name = name
        self.country = country
        self.days = days


class Person:
    
    def __init__(self, name, age, occupation = 'General Citizen'):
        self.name = name
        self.age = age
        self.occupation = occupation
        self.dose = 0
        self.first_dose = 'Not Given'
        self.second_dose = 'Not Given'
    
    def pushVaccine(self, vaccine, dose = '1st Dose'):
        if self.age <= 25 and self.occupation != 'Student':
            print(f'Sorry {self.name}, Minimum age for taking vaccines is 25 years now.')
        else:
            if self.dose < 2:
                if dose == '1st Dose':
                    self.vaccine_name = vaccine.name
                    self.first_dose = 'Given'
                    self.second_dose = f'Please come after {vaccine.days} days'
                    self.dose += 1
                    print(f'1st dose done for {self.name}')
                else:
                    if self.vaccine_name == vaccine.name:
                        self.second_dose = 'Given'
                        self.dose += 1
                        print(f'2nd dose done for {self.name}')
                    else :
                        print(f'Sorry {self.name}, you can’t take 2 different vaccines')
            else:
                print(f'{self.name} has already takes 2 doses')
    
    def showDetail(self):
        print(f'Name: {self.name} Age: {self.age} Type: {self.occupation}')
        print(f'Vaccine name: {self.vaccine_name}')
        print(f'1st dose: {self.first_dose}')
        print(f'2nd dose: {self.second_dose}')


astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
