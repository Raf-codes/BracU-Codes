```python
#1
class Student:
    ID = 0

    def __init__(self, n, dept, age, cgpa):
        self.n = n
        self.dept = dept
        self.age = age
        self.cgpa = cgpa

    def get_details(self):
        Student.ID += 1
        print("ID:", Student.ID)
        print("Name:", self.n)
        print("Department:", self.dept)
        print("Age:", self.age)
        print("CGPA:", self.cgpa)

    @classmethod
    def from_string(cls, info):
        n, dept, age, cgpa = info.split('-')
        return cls(n, dept, age, cgpa)
         


s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_string("Sumaiya-BBA-23-3.96")
s4.get_details()
print("-----------------------")
```

```python
#2
class Assassin:
    count = 0
    def __init__(self, name, rate):
        self.name = name
        self.rate = rate
        Assassin.count += 1

    def printDetails(self):
        print('Name:', self.name)
        print('Success rate: ' + str(self.rate) + '%')
        print('Total number of Assassin:', Assassin.count)

    @classmethod
    def failureRate(cls, name, rate):
        return cls(name, 100 - int(rate))
    @classmethod
    def failurePercentage(cls, name, rate):
        return cls(name, 100 - int(rate))



john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
#5

class Employee:
    def __init__(self, name, wp):
        self.name = name
        self.workingPeriod = wp

    @classmethod
    def employeeByJoiningYear(cls, name, year):
        wp = 2022 - year
        return cls(name, wp)

    @staticmethod
    def experienceCheck(wp, gender):
        if wp < 3:
            if gender == "male":
                return "He is not experienced"
            else:
                return "She is not experienced"
        else:
            if gender == "male":
                return "He is experienced"
            else:
                return "She is experienced"


employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#7

class Cat:
    Number_of_cats = 0
    doing = "sitting"
    color = "white"

    def __init__(self, cc, cd):
        self.color = cc
        self.doing = cd
        Cat.Number_of_cats += 1

    @classmethod
    def no_parameter(cls):
        return cls(cls.color, cls.doing)


    @classmethod
    def first_parameter(cls, color):
        cls.color = color
        return cls(cls.color, cls.doing)


    @classmethod
    def second_parameter(cls, action):
        cls.doing = action
        return cls("Grey", cls.doing)


    def printCat(self):
        print(self.color, " cat is ", self.doing)

    def changeColor(self, nc):
        self.color = nc

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#8
import math
class Cylinder:
    radius = 5
    height = 18

    def __init__(self, r, h): 
        print(f"Default radius={Cylinder.radius} and height={Cylinder.height}")
        Cylinder.radius = r
        Cylinder.height = h
        print(f'Updated radius={Cylinder.radius} and height={Cylinder.height}')
    @classmethod
    def swap(cls,r,h):
      return cls(h,r)
    @classmethod
    def changeFormat(cls,a):
      r1,h1=a.split("-")
      return cls(float(r1),float(h1))
    @staticmethod
    def volume(r, h):
        print(f"volume: {math.pi*r*r*h}")

    @staticmethod
    def area(r, h):
      print(f"Area: {float(2*math.pi*r*r)+float(2*math.pi*r*h)}")

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#10
class SultansDine:
    total_branch = 0
    total_sell = 0
    branch_list = []

    def __init__(self, branch):
        self.branch = branch
        self.sell = 0

        SultansDine.total_branch += 1
        
    
    def sellQuantity(self, quantity):
        if quantity < 10:
            self.sell = quantity * 300
        elif quantity < 20:
            self.sell = quantity * 350
        else:
            self.sell = quantity * 400

        SultansDine.total_sell += self.sell
        SultansDine.branch_list.append([self.branch, self.sell])
    
    def branchInformation(self):
        print('Branch Name:', self.branch)
        print(f'Branch Sell: {self.sell} Taka')
    
    @classmethod
    def details(cls):
        print('Total Number of branch(s):', cls.total_branch)
        print('Total Sell:', cls.total_sell)

        if cls.total_branch != 0:
            for i in cls.branch_list:
                name = i[0]
                sell = i[1]
                rate = (sell / cls.total_sell) * 100

                print(f'Branch Name: {name}, Branch Sell: {sell} Taka')
                print(f'Branch consists of total sell\'s: {round(rate, 2)}%')


SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
