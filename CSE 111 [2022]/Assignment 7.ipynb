```python
#2
class Vehicle:

    def __init__(self):
        self.x = 0
        self.y = 0
    def moveUp(self):
        self.y += 1
    def moveDown(self):
        self.y -= 1
    def moveRight(self):
        self.x += 1
    def moveLeft(self):
        self.x -= 1
    def __str__(self):
        return '('+str(self.x)+' , '+str(self.y)+')'

#write your code here
class Vehicle2010(Vehicle):
    def __init__(self):
        super().__init__()
    def moveLowerLeft(self):
        self.moveDown()
        self.moveLeft()
    def moveLowerRight(self):
        self.moveDown()
        self.moveRight()
    def moveUpperLeft(self):
        self.moveUp()
        self.moveLeft()
    def moveUpperRight(self):
        self.moveUp()
        self.moveRight()
    def equals(self,o):
        return True if self.x==o.x and self.y==o.y else False
            

print('Part 1')
print('------')
car = Vehicle()
print(car)
car.moveUp()
print(car)
car.moveLeft()
print(car)
car.moveDown()
print(car)
car.moveRight()
print(car)
print('------')
print('Part 2')
print('------')
car1 = Vehicle2010()
print(car1)
car1.moveLowerLeft()
print(car1)
car2 = Vehicle2010()
car2.moveLeft()
print(car1.equals(car2))
car2.moveDown()
print(car1.equals(car2))
```

```python
#3
class Tournament:
 def __init__(self,name='Default'):
  self.__name = name
 def set_name(self,name):
  self.__name = name
 def get_name(self):
  return self.__name
#write from here


class Cricket_Tournament(Tournament):
  def __init__(self,name='Default',team_no=0,type='No type'):
   super().set_name(name)
   self.team_no=team_no
   self.type=type
  def detail(self):
    return 'Cricket Tournament Name: '+super().get_name()+'\nNumber of Teams: '+str(self.team_no)+'\nType: '+self.type
    
class Tennis_Tournament(Tournament):
  def __init__(self,name,player_no):
    super().set_name(name)
    self.player_no=player_no
  def detail(self):  
    return 'Tennis Tournament Name: '+super().get_name()+'\nNumber of Players: '+str(self.player_no)

ct1 = Cricket_Tournament()
print(ct1.detail())
print("-----------------------")
ct2 = Cricket_Tournament("IPL",10,"t20")
print(ct2.detail())
print("-----------------------")
tt = Tennis_Tournament("Roland Garros",128)
print(tt.detail())
```

```python
#4
class Product:
 def __init__(self,id, title, price):
  self.__id = id
  self.__title = title
  self.__price = price

 def get_id_title_price(self):
  return "ID: "+str(self.__id)+" Title: "+self.__title+ " Price: "+str(self.__price)

#write from here

class Book(Product):
  def __init__(self,id,title,price,isbn,publisher):
    super().__init__(id,title,price)   
    self.isbn=isbn
    self.publisher=publisher

  def printDetail(self):
    return self.get_id_title_price() +'\nISBN: '+self.isbn+ ' Publisher: '+self.publisher

class CD(Product):
  def __init__(self,id,title,price,band,time,genre):
    super().__init__(id,title,price)
    self.band=band
    self.time=time
    self.genre=genre
    
  def printDetail(self):
    return self.get_id_title_price() +'\nBand '+self.band+ ' Duration: '+str(self.time)+'minutes' + '\nGenre: '+self.genre
    
book = Book(1,"The Alchemist",500,"97806","HarperCollins")
print(book.printDetail())
print("-----------------------")
cd = CD(2,"Shotto",300,"Warfaze",50,"Hard Rock")
print(cd.printDetail())
```

```python
#task 6
class Shape:
    def __init__(self, name='Default', height=0, base=0):
        self.area = 0
        self.name = name
        self.height = height
        self.base = base

    def get_height_base(self):
        return "Height: " + str (self.height) + ",Base: " + str (self.base)

#write from here
class triangle (Shape):
    def __init__(self, name='Default', height=0, base=0):
        super ().__init__ (name, height, base)
        self.area = 0

    def calcArea(self):
        self.area = self.height * self.base * 0.5
        return self.area

    def printDetail(self):
        a = f'Shape name: {self.name}  \n'
        b= self.get_height_base ()
        c= '\n' 'Area:' + str (self.area)
        return a+b+c


class trapezoid (Shape):
    def __init__(self, name, height, base, side_a):
        super ().__init__ (name, height, base)
        self.side_a = side_a
        self.area = 0

    def calcArea(self):
        self.area = ((self.side_a + self.base) * 0.5) * self.height
        return self.area

    def printDetail(self):
        d = f'Shape name: {self.name} \n'
        e= self.get_height_base ()
        f= f' Side_A: {self.side_a}'
        g= '\nArea: ' + str (self.area)
        return d+e+f+g


tri_default = triangle ()
tri_default.calcArea ()
print (tri_default.printDetail ())
print ('--------------------------')
tri = triangle ('Triangle', 10, 5)
tri.calcArea ()
print (tri.printDetail ())
print ('---------------------------')
trap = trapezoid ('Trapezoid', 10, 6, 4)
trap.calcArea ()
print (trap.printDetail ())
```

```python
class SportsPerson:
  def __init__(self, team_name, name, role):
    self.__team = team_name
    self.__name = name
    self.role = role
    self.earning_per_match = 0


  def get_name_team(self):
    return 'Name: '+self.__name+', Team Name: ' +self.__team
#write from here


class Player(SportsPerson):
  def __init__(self,team_name, name, role,goal,m_played):
    super().__init__(team_name, name, role)
    self.goal=goal
    self.m_played=m_played
    self.earning_per_match = (goal * 1000) + (m_played * 10)
  
  def calculate_ratio(self):
    self.g_ratio=self.goal/self.m_played
  
  def print_details(self):
    print(self.get_name_team())
    print('Team Role:',self.role)
    print(f'Total Goal: {self.goal}, Total Played: {self.m_played}')
    print('Goal Ratio:',self.g_ratio)
    print('Match Earning:',self.earning_per_match)


class Manager(SportsPerson):
    def __init__(self,team_name, name, role,match):
      super().__init__(team_name, name, role)
      self.match=match
      self.earning_per_match= match*1000
    
    def print_details(self):
      print(self.get_name_team())
      print('Team Role:',self.role)
      print('Total Win:',self.match)
      print('Match Earning:',self.earning_per_match)





player_one = Player('Juventus', 'Ronaldo', 'Striker', 25, 32)
player_one.calculate_ratio()
player_one.print_details()
print('------------------------------------------')
manager_one = Manager('Real Madrid', 'Zidane', 'Manager', 25)
manager_one.print_details()
```
