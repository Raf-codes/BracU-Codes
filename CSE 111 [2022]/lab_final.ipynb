```python
class Train:
    def __init__(self, name=None):
        self.name = name
        self.speed = 100
        self.__totalSeat = 0
        self.hasNightCoach = False

    def getSeatStatus(self):
        return self.__totalSeat

    def setSeatStatus(self, num):
        self.__totalSeat = num

    def printDetails(self):
        print (f'{self.name} is currently running on {self.speed}mph.')


class Passenger:
    def checkSeat(self, train):
        print ("=========================")
        if train.getSeatStatus () > 0:
            print ("Seat available.")
        else:
            print ("Seat not available.")


# WRITE YOUR CODE HERE
class BulletTrain(Train):
    total_dest=0
    def __init__(self,name):
        super().__init__(name)
        self.speed=500
        self.rout = []
        self.string = ""
        self.hasNightCoach=True

    def printRoute(self):
        if len(self.rout)==0:
            print(f"No rout found for {self.name}")
        else:
            print(self.string[0:-3])


    def createSeat(self,num):
        super().setSeatStatus(num)


    def setRoute(self,*destination):
        self.destination=destination
        for i in self.destination:
            self.rout.append(i)
            BulletTrain.total_dest+=1

        for i in self.rout:
            self.string+=i + "-->"

    def printDetails(self):
        super().printDetails()
        print(f'Total seat remaining {super().getSeatStatus()}')
        if self.hasNightCoach==True:
            print("Newly added night coach")


t = Train ('Rajdhani Express')
t.printDetails ()
b1 = BulletTrain ('Bullet SP1876')
b1.printRoute ()
print ('Total destinations covered: ', BulletTrain.total_dest)
p = Passenger ()
p.checkSeat (b1)
b1.createSeat (20)
p.checkSeat (b1)
print ("------------------------")
b1.setRoute ('Mohakhali', 'Badda')
print ('Total destinations covered: ', BulletTrain.total_dest)
b1.printRoute ()
print ("------------------------")
b1.setRoute ('Gulshan', 'Baridhara')
print ('Total destinations covered: ', BulletTrain.total_dest)
b1.printRoute ()
print ("------------------------")
b1.printDetails ()
print ("------------------------")
b2 = BulletTrain ('Bullet PB2190')
b2.printRoute ()
p.checkSeat (b2)
print ("------------------------")
b2.hasNightCoach = False
b2.createSeat (30)
b2.printDetails ()
print ("------------------------")
b2.setRoute ('Rajshahi')
b2.printRoute ()
p.checkSeat (b2)
print ('Total destinations covered: ', BulletTrain.total_dest)
```
