```python
#2022 srping mid
class RickMote:
    def __init__(self):
        self._is_on = False
        self.ch1=6
        self.v2=3
    def power(self):
        if not self._is_on:
            self._is_on = True
        else:
            self._is_on = False
    def changeChannel(self,v1=3):
        self.ch1=v1
        if self.ch1>10:
            print("ERROR! Tv channel doesn't exist!")
            self.ch1=3
        else:
            self.ch1=v1


    def changeVolumeLevel(self,v2=None):
        if v2==None:
            self.v2+=1
        else:
            self.v2+=v2
            if self.v2>20:
               print("Can't change volume anymore!")
               self.v2=6





    def showInfo(self):
        if self._is_on == True:
            print("ID Cable Box Status:")
            if self._is_on == True:
                print ("Cable Box is : ON")
            else:
                print ("Cable Box is : OFF")
            print("channel is:",self.ch1)
            print("volume is:",self.v2)
        else:
            print ("ID Cable Box Status:")
            print ("Cable Box is : OFF")


oTV = RickMote()
oTV.power()
oTV.showInfo()
print("###############################")
oTV.changeChannel()
oTV.changeChannel(3)
oTV.changeVolumeLevel()
oTV.showInfo()
print("###############################")
oTV.power()
oTV.showInfo()
print("###############################")
oTV.power()
oTV.changeVolumeLevel(4)
oTV.showInfo()
print("###############################")
oTV. changeVolumeLevel(-2)
oTV.showInfo()
print("###############################")
oTV. power()
oTV.changeVolumeLevel(10)
oTV.changeChannel(9)
oTV.changeVolumeLevel(-1)
oTV.showInfo()
print("###############################")
oTV.power()
oTV.changeChannel(11)
oTV.changeVolumeLevel(10)
oTV.showInfo()
```
