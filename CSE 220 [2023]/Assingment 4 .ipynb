**Assignment Part 1: Doubly Linked List**

For Assignment Part 1, you should write your full code in the following cells along with your driver codes on your own.

```python
#Assignment Part 1
class Patient:
    def __init__(self, id, name, age, bloodgroup):
        self.id = id
        self.name = name
        self.age = age
        self.bloodgroup = bloodgroup
        self.next = None
        self.prev = None
```

```python
class WRM:
    def __init__(self):
        self.head = None
        self.tail = None

    def is_empty(self):
        return self.head is None

    def register_patient(self, id, name, age, bloodgroup):
        new_patient = Patient(id, name, age, bloodgroup)

        if self.is_empty():
            self.head = new_patient
            self.tail = new_patient
        else:

            self.tail.next = new_patient
            new_patient.prev = self.tail
            self.tail = new_patient

        print(f"Patient {name} successfully registered.")

    def serve_patient(self):
        if self.is_empty():
            print("No patient to serve.")
            return None

        served_patient = self.head
        self.head = self.head.next

        if self.head:
            self.head.prev = None

        return served_patient

    def cancel_all(self):
        self.head = None
        self.tail = None
        print("All appointments canceled.")

    def can_doctor_go_home(self):
        return self.is_empty()

    def show_all_patients(self):
        if self.is_empty():
            print("No patient to serve.")
            return

        current_patient = self.head
        while current_patient:
            print(f"Patient name: {current_patient.name}")
            current_patient = current_patient.next


```

```python
#Write a Tester Code in this cell
print("**Welcome to Waiting Room Management System**")
wrm = WRM()

while True:
    print("\n1. Register Patient")
    print("2. Serve a Patient")
    print("3. Show All Patients")
    print("4. Can Doctor Go Home?")
    print("5. Cancel All Appointments")
    print("6. Exit")

    try:
        choice = int(input("Enter your choice: "))
    except ValueError:
        print("Invalid input. Please enter a number between 1 and 6.")
        continue

    if choice == 1:
        id = int(input("Enter patient ID: "))
        name = input("Enter patient name: ")
        age = int(input("Enter patient age: "))
        bloodgroup = input("Enter patient blood group: ")
        wrm.register_patient(id, name, age, bloodgroup)

    elif choice == 2:
        served_patient = wrm.serve_patient()
        if served_patient:
            print(f"Patient {served_patient.name} is served.")

    elif choice == 3:
        wrm.show_all_patients()

    elif choice == 4:
        if wrm.can_doctor_go_home():
            print("Yes")
        else:
            print("No")

    elif choice == 5:
        wrm.cancel_all()

    elif choice == 6:
        print("Thank you for using the Waiting Room Management System :)")
        break

    else:
        print("Invalid choice. Please enter a number between 1 and 6.")
```

**Assignment Part 2: Stack**

Linked List based Stack is implemented in the following cell.

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```

Task 1: Diamond Count

```python
def diamond_count(stack,string):
  count=0
  for i in range(len(string)):
    if string[i]=="<":
      stack.push(string[i])
    elif string[i] == ">":
      if stack.peek()=="<":
        stack.pop()

        count+=1

  return count



print('Test 01')
stack = Stack()
string = '<..><.<..>> '
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')


print('Test 02')
stack = Stack()
string = '<<<..<......<<<<....>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 1
unittest.output_test(returned_value, 1)
print('-----------------------------------------')


print('Test 03')
stack = Stack()
string = '>>><...<<..>>...>...>>>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')
```

Task 2: Tower of Blocks

```python
def remove_block(stack, n):


    if n < 1:
        print("Invalid index")
        return None

    temp_stack = Stack()

    counter = 1
    while not stack.isEmpty() and counter < n:
        temp_stack.push(stack.pop())
        counter += 1

    removed_block = stack.pop()

    while not temp_stack.isEmpty():
        stack.push(temp_stack.pop())

    return removed_block

# Driver code
print('Test 01')
st = Stack()
st.push(4)
st.push(19)
st.push(23)
st.push(17)
st.push(5)
print('Stack:')
print_stack(st)  # Replace with your print_stack function
print('------')
removed_block = remove_block(st, 2)
if removed_block is not None:
    print('Removed block:', removed_block)
print('After Removal:')
print_stack(st)
print('------')

print()
print('======================================')
print()

print('Test 02')
st = Stack()
st.push(73)
st.push(85)
st.push(15)
st.push(41)
print('Stack:')
print_stack(st)  # Replace with your print_stack function
print('------')
removed_block = remove_block(st, 3)
if removed_block is not None:
    print('Removed block:', removed_block)
print('After Removal:')
print_stack(st)
print('------')
```

Task 3: Stack Reverse

```python
def conditional_reverse(stack) :


    reversed_stack = Stack()
    previous_element = None

    while not stack.isEmpty():
        current_element = stack.pop()

        if current_element != previous_element:
            reversed_stack.push(current_element)

        previous_element = current_element

    return reversed_stack


print('Test 01')
st=Stack()
st.push(10)
st.push(10)
st.push(20)
st.push(20)
st.push(30)
st.push(10)
st.push(50)
print('Stack:')
print_stack(st)
print('------')
reversed_stack=conditional_reverse(st)
print('Conditional Reversed Stack:')
print_stack(reversed_stack) # This stack contains 50, 10, 30, 20, 10 in this order whereas top element should be 10
print('------')
```
