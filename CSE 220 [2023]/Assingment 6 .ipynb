```python
class vehicleNodes:
  def __init__(self, brand, vehicle_type, rent, passenger, next = None):
      self.brand = brand
      self.vehicle_type = vehicle_type
      self.rent = rent
      self.passenger = passenger
      self.next = next
```

```python
class VehicleHashTable:
  def __init__(self, size):
    self.vehicleTable = [None]*size
    self.empty_slot = {}
    for i in range(size):
      self.empty_slot[i] = False

  def create_from_vehicle_info_array(self, arr):
    for i in arr:
      self.insert_vehicle(i)


  def print_vehicle_hashtable(self):
    idx = 0
    for i in self.vehicleTable:
      print(idx, ':', end = ' ')
      head = i
      while head != None:
        print(f'(Brand: {head.brand}, Type: {head.vehicle_type}, Rent: {head.rent}, No. of Passengers: {head.passenger})', end = '---->')
        head = head.next
      print('None')
      print()
      idx += 1
  def calculate_idx(self,brand):
      total= sum(ord(i) for i in brand)
      total= total % len(self.vehicleTable)
      return total


  def find_empty_slot(self):
    for k,v in self.empty_slot.items():
      idx = k
      del self.empty_slot[k]
      return idx



  def __hash_function(self, brand):
      index = self.calculate_idx(brand)

      if index in self.empty_slot:
        del self.empty_slot[index]
      return index



  def insert_vehicle(self, vehicle):
    index = self.__hash_function(vehicle[0])
    if self.vehicleTable[index] == None:
      new_node= vehicleNodes(vehicle[0],vehicle[1],vehicle[2],vehicle[3])
      self.vehicleTable[index]=new_node
    else:
          if vehicle[0] == self.vehicleTable[index].brand:
            head = self.vehicleTable[index]
            head.next = vehicleNodes(vehicle[0],vehicle[1],vehicle[2],vehicle[3])

          else:
            index = self.find_empty_slot()

            new_node= vehicleNodes(vehicle[0],vehicle[1],vehicle[2],vehicle[3])
            self.vehicleTable[index]=new_node
    return self.vehicleTable
```

```python
class Node_pair:
  def __init__(self, key, value, next = None):
    self.key, self.value, self.next = key, value, next


class Hashtable:
  def __init__(self, size):
    self.ht = [None]*size


  def insert(self, s):
    if self.search_hashtable(s) == 'Found':
      print(s,'Already Inserted. Cannot reinsert.')
      print('===============================')
      return
    hashed_index = self.__hash_function(s[0])
    pair = Node_pair(s[0], s[1])
    if self.ht[hashed_index] == None:
      self.ht[hashed_index] = pair
    else:
      pair.next = self.ht[hashed_index]
      self.ht[hashed_index] = pair


  def create_from_array(self, arr):
    for i in arr:
      self.insert(i)

  def print_hashtable(self):
    idx = 0
    for i in self.ht:
      print(idx, ':', end = ' ')
      head = i
      while head != None:
        print(f'(key: {head.key}, value: {head.value})', end = '-->')
        head = head.next
      print('None')
      print()
      idx += 1

  def __hash_function(self, key):
        if len(key) % 2 != 0:
            key += 'N'
        total = sum(int(''.join(map(str, [ord(char) for char in key[i:i+2]]))) for i in range(0, len(key), 2))
        return total % len(self.ht)

  def search_hashtable(self, s):
        hashed_index = self.__hash_function(s[0])
        current = self.ht[hashed_index]

        while current is not None:
            if current.key == s[0] and current.value == s[1]:
                return 'Found'
            current = current.next

        return 'Not Found'
```

Task 3

Implement the __hash_function() and insert() methods

```python
class Node:
    def __init__(self, value=None, next=None):
        self.value = value
        self.next = next


class HashTable:
    def __init__(self, length):
        n = Node()
        self.ht = [n] * length
        self.length = length

    def show(self):
        count = 0
        for i in self.ht:
            temp = i
            print(count, end=" ")
            while temp != None:
                print(temp.value, end="-->")
                temp = temp.next
            count += 1
            print()

    def __hash_function(self, key):
        length = len(key)
        sum = 0
        if length % 2 == 0:
            for i in range(0, length, 2):
                sum += ord(key[i])
        else:
            for i in range(1, length, 2):
                sum += ord(key[i])
        return sum % self.length

    def insert(self, key, value):
        index = self.__hash_function(key)
        node = Node((key, value))

        if self.ht[index].value is None:
            self.ht[index] = node
        else:
            current = self.ht[index]
            if current.value[1] < value:
                node.next = current
                self.ht[index] = node
            else:
                while current.next is not None and current.next.value[1] >= value:
                    current = current.next
                temp = current.next
                current.next = node
                node.next = temp
```

```python
#Driver Code
ht = HashTable(3)


ht.insert("apple", 20)
ht.insert("coconut", 90)
ht.insert("cherry", 50)
ht.show()
print("------------")
ht.insert("banana", 30)
ht.insert("pineapple", 50)
ht.show()

# Driver Code Output:
# 0 ('coconut', 90)-->
# 1 ('apple', 20)-->
# 2 ('cherry', 50)-->
# ------------
# 0 ('coconut', 90)-->('pineapple', 50)-->('banana', 30)-->
# 1 ('apple', 20)-->
# 2 ('cherry', 50)-->
```

```python
class Node_pair:
  def __init__(self, key, value, next = None):
    self.key, self.value, self.next = key, value, next


class Hashtable:
  def __init__(self, size):
    self.ht = [None]*size


  def insert(self, s):
    hashed_index = self.__hash_function(s[0])
    pair = Node_pair(s[0], s[1])
    if self.ht[hashed_index] == None:
      self.ht[hashed_index] = pair
    else:
      pair.next = self.ht[hashed_index]
      self.ht[hashed_index] = pair


  def create_from_array(self, arr):
    for i in arr:
      self.insert(i)

  def print_hashtable(self):
    idx = 0
    for i in self.ht:
      print(idx, ':', end = ' ')
      head = i
      while head != None:
        print(f'({head.key}, {head.value})', end = '-->')
        head = head.next
      print('None')
      idx += 1

  def __hash_function(self, key):
    return (key + 3) % 6

  def remove(self, key):
    hashed_index = self.__hash_function(key)
    prev = None
    curr = self.ht[hashed_index]

    while curr is not None:
      if curr.key == key:
        if prev is None:
          self.ht[hashed_index] = curr.next
        else:
          prev.next = curr.next
        return

      prev = curr
      curr = curr.next
```
