```python
class BTNode:
  def __init__(self, elem):
    self.elem = elem
    self.right = None
    self.left = None
```

```python
def inorder(root):
  if root == None:
    return

  inorder(root.left)
  print(root.elem, end = ' ')
  inorder(root.right)

```

```python
def tree_construction(arr, i = 1):
  if i>=len(arr) or arr[i] == None:
    return None
  p = BTNode(arr[i])
  p.left = tree_construction(arr, 2*i)
  p.right = tree_construction(arr, 2*i+1)
  return p


root2 = tree_construction([None, 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', None, None, None, 'I', 'J', None, 'k'])
inorder(root2)
```

```python
def smallest_level(root):
    level_mins = {}
    traverse_tree(root, level_mins, 0)
    return level_mins

def traverse_tree(node, level_mins, level):
    if not node:
        return

    if level not in level_mins:
        level_mins[level] = node.elem
    else:
        level_mins[level] = min(level_mins[level], node.elem)

    traverse_tree(node.left, level_mins, level + 1)
    traverse_tree(node.right, level_mins, level + 1)


#DRIVER CODE
root = tree_construction([None, 4,9,2,3,-5,None,7])
print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root) #Given Tree Inorder Traversal:  3 9 5 4 2 7
print()
print('Level Wise Smallest Value: ', end = ' ')
print(smallest_level(root)) #Level Wise Smallest Value:  {0: 4, 1: 2, 2: -5}
```

```python
def inorder_predecessor(root, x):
    if root is None:
        return None

    if root.right is None and root.elem < x.elem:
        return root

    predecessor = None
    curr = root

    while curr is not None:
        if curr.elem < x.elem:
            predecessor = curr
            curr = curr.right
        else:
            curr = curr.left

    return predecessor
#DRIVER CODE
root = BTNode(20)
n1 = BTNode(8)
n2 = BTNode(22)
n3 = BTNode(4)
n4 = BTNode(12)
n5 = BTNode(10)
n6 = BTNode(14)

root.left = n1
root.right = n2

n1.left = n3
n1.right = n4

n4.left = n5
n4.right = n6

print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root) #Given Tree Inorder Traversal:  4 8 10 12 14 20 22
print()

x = root
print(f'Inorder predecessor of node {x.elem}: {inorder_predecessor(root, x).elem}') #Inorder predecessor of node 20: 14
```

```python
def LCA(root, x, y):
    if root is None:
        return None
    if x > root.elem and y > root.elem:
        return LCA(root.right, x, y)
    elif x < root.elem and y < root.elem:
        return LCA(root.left, x, y)
    else:
        return root


#DRIVER CODE
root = BTNode(15)
n1, n2 = BTNode(10), BTNode(25)
n3, n4 = BTNode(8), BTNode(12)
n5, n6 = BTNode(6), BTNode(9)
n7, n8 = BTNode(20), BTNode(30)
n9, n10 = BTNode(18), BTNode(22)

root.left, root.right = n1, n2
n1.left, n1.right = n3, n4
n3.left, n3.right = n5, n6
n2.left, n2.right = n7, n8
n7.left, n7.right = n9, n10

cases = [(6, 12), (20, 6), (18, 22), (20, 25), (10, 12)]

for x, y in cases:
    print(f"LCA({x}, {y}) =", LCA(root, x, y).elem)
```

```python
def swap_child(root, level, M):
    if root is None or level == M:
        return root


    root.left, root.right = swap_child(root.right, level + 1, M), swap_child(root.left, level + 1, M)

    return root


# Driver Code
root = BTNode('A')
root.left = BTNode('B')
root.right = BTNode('C')
root.left.left = BTNode('D')
root.left.right = BTNode('E')
root.right.right = BTNode('F')
root.left.left.left = BTNode('G')
root.left.left.right = BTNode('H')
root.left.right.left = BTNode('I')
root.right.right.left = BTNode('J')


print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root)   #Given Tree Inorder Traversal: G D H B I E A C J F
print()

root2 = swap_child(root, 0, 2)
print('Swapped Tree Inorder Traversal: ', end = ' ')
inorder(root2)  #Swapped Tree Inorder Traversal: J F C A I E B G D H
```
