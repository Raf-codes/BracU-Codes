```python
#run hoite onk time lagbe eta 
from sympy import*
import numpy as np
import matplotlib.pyplot as plt
x= symbols('x')
f = 2*x**16 - 4*x**2+ (1/x**3)
df= diff(f,x) #1st differentiation
df2=diff(df,x) #2nd differentiation
df3=diff(df2,x) #3rd differentiation
print("1st differentiation",df)
print("2nd differentiation",df2)
print("3rd differentiation",df3)
p_x=np.arange(-5,6)
f = lambdify(x, df3)
y=f(p_x)
plt.plot(p_x,y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('2*x**16 - 4*x**2+ (1/x**3)')
plt.show()
```

```python
from sympy import*
import numpy as np
import matplotlib.pyplot as plt
x= symbols('x')
f = 2*exp(4*x)-9*x**2
df= diff(f,x)
df2=diff(df,x)
df3=diff(df2,x)
print("1st differentiation",df)
print("2nd differentiation",df2)
print("3rd differentiation",df3)
p_x=np.arange(-5,6)
f = lambdify(x, f)
y=f(p_x)
plt.plot(p_x,y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('2*exp(4*x)-9*x**2')
plt.show()
```

```python
import sympy as smp
x,a,b = smp.symbols('x a b')
f = smp.sin(a*x)*smp.exp(-b*x)
smp.integrate(f, x)
smp.integrate(f,x).simplify() # for simplifying the answer
```

```python
import sympy as smp


x = smp.symbols('x')

y = smp.Function('y')

y1 = smp.Derivative(y(x), x)

y2 = smp.Derivative(y(x), x, x)
eqdiff = y1 + y(x) * smp.tan(x) - smp.exp(2 * x) * smp.cos(x)


smp.dsolve(eqdiff, y(x))
x0 = 0
smp.dsolve(eqdiff, y(x), ics={y(x0): 2, y(x).diff(x).subs(x, x0): 0})
```
